```python
import re
from google.colab import files

# 上传多个 RIS 文件
uploaded = files.upload()

# 提取标题和摘要的函数
def extract_titles_and_abstracts_from_ris(files):
    titles = []  # 存储提取的标题
    abstracts = []  # 存储提取的摘要

    # 正则表达式模式，用于匹配标题 (TI) 和摘要 (AB)
    title_pattern = r'^TI\s*-\s*(.*)$'  # 匹配以 'TI  - ' 开头的标题行
    abstract_pattern = r'^AB\s*-\s*(.*)$'  # 匹配以 'AB  - ' 开头的摘要行

    # 遍历上传的文件
    for filename in uploaded.keys():
        print(f"Processing file: {filename}")

        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()  # 读取文件内容

            # 找到所有匹配的标题和摘要
            titles_in_file = re.findall(title_pattern, content, re.MULTILINE)
            abstracts_in_file = re.findall(abstract_pattern, content, re.MULTILINE)

            # 将提取到的标题和摘要分别添加到列表
            titles.extend(titles_in_file)
            abstracts.extend(abstracts_in_file)

    return titles, abstracts

# 提取标题和摘要
titles, abstracts = extract_titles_and_abstracts_from_ris(uploaded)

# 打印部分结果查看是否正确
print("Some extracted titles:")
print(titles[:5])  # 打印前五个标题
print("Some extracted abstracts:")
print(abstracts[:5])  # 打印前五个摘要

# 保存标题到 txt 文件
with open('/content/titles.txt', 'w', encoding='utf-8') as title_file:
    for title in titles:
        title_file.write(title + '\n')

# 保存摘要到 txt 文件
with open('/content/abstracts.txt', 'w', encoding='utf-8') as abstract_file:
    for abstract in abstracts:
        abstract_file.write(abstract + '\n')

# 提示文件保存完毕
print("标题和摘要已保存到 titles.txt 和 abstracts.txt")

# 提供下载链接
files.download('/content/titles.txt')
files.download('/content/abstracts.txt')
```

```python
import re
from collections import Counter
from google.colab import files
import os

# 读取txt文件内容
def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

# 词频统计函数
def word_frequency(text):
    # 将文本转为小写并移除非字母数字字符（保留空格和字母数字）
    text = re.sub(r'[^a-zA-Z\s]', '', text.lower())  # 去除标点符号并转换为小写
    words = text.split()  # 按空格分割文本为单词
    return Counter(words)  # 返回每个单词的词频

# 输出词频统计结果
def print_word_frequency(file_name, counter):
    print(f"Word frequency for {file_name}:")
    # 按照词频从高到低排序并打印
    for word, freq in counter.most_common():
        print(f'{word}: {freq}')
    print("\n")

# 获取子文件夹路径
folder_path = '/content/123'  # 123是子文件夹名称，请确保文件夹路径正确

# 确保文件夹路径存在
if not os.path.exists(folder_path):
    print(f"文件夹 {folder_path} 不存在。")
else:
    # 获取文件夹中的txt文件
    files_in_folder = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

    if len(files_in_folder) != 2:
        print("文件夹中未找到两个txt文件。")
    else:
        # 获取两个txt文件的路径
        file_path_1 = os.path.join(folder_path, files_in_folder[0])
        file_path_2 = os.path.join(folder_path, files_in_folder[1])

        # 读取文件内容
        text1 = read_file(file_path_1)
        text2 = read_file(file_path_2)

        # 获取词频统计
        counter1 = word_frequency(text1)
        counter2 = word_frequency(text2)

        # 打印词频统计
        print_word_frequency(file_path_1, counter1)
        print_word_frequency(file_path_2, counter2)
```

```python
# 保存词频统计结果到文件
def save_to_file(counter, output_file):
    with open(output_file, 'w', encoding='utf-8') as file:
        for word, freq in counter.most_common():
            file.write(f'{word}: {freq}\n')

# 保存结果到文件
save_to_file(counter1, '/content/123/file1_word_frequency.txt')
save_to_file(counter2, '/content/123/file2_word_frequency.txt')

print("词频统计结果已保存到文件")
```
